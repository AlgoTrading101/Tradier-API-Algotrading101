```python
url = base_url + 'v1/markets/timesales'

response = requests.get(url,
                        params={'symbol': 'V', 'interval': '1min', 'start': '2020-12-23 15:00', 'end': '2020-12-23 16:00', 'session_filter': 'all'},
                       headers = headers
                       )
json_response = response.json()
json_response['series']['data']
```

```python
df = pd.DataFrame(json_response['series']['data'], columns =['price'])
df
```

```python
max_price = df.price.max()
min_price = df.price.min()

url = base_url + 'v1/accounts/{account_id}/orders'

if df.price.iloc[-1] < max_price * 0.99:
    try:
        response = requests.post(url,
                                params={'class' : 'equity', 'symbol' : 'MA', 'side' : 'buy', 'quantity' : '500.00', 'type' : 'market', 'duration' : 'day' },
                                headers = headers)
        print ('Dropped 1% and buy order was placed, Current price: {} Max price: {}'.format(df.price.iloc[-1], max_price))
        
    except Exception as e:
            print('Error occured while placing order:',e)
            
elif df.price.iloc[-1] > min_price * 1.01:
    try:
        response = requests.post(url,
                                params={'class' : 'equity', 'symbol' : 'MA', 'side' : 'sell', 'quantity' : '500.00', 'type' : 'market', 'duration' : 'day' },
                                headers = headers)
        print ('Risen 1% and sell order was placed, Current price: {} Min price: {}'.format(df.price.iloc[-1], min_price))
    
    except Exception as e:
            print('Error occured while placing order:',e)
    
else:
    print('No order was placed, Current price: {} Min price: {} Max price: {}\n'.format(df.price.iloc[-1], min_price, max_price))  
```
